```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statistics as st
import math
```

```python
data_path = 'steady.csv'

df = pd.read_csv(data_path)
```

```python
trend = 'S1-PT-1001A'

y = df[trend]
x = df['Seconds']
c= np.random.random(len(y))
```

```python
spline_model = make_interp_spline(x, y, k=5)
```

```python
x_spline = np.linspace(x.min(), x.max(), 150)
y_spline = spline_model(x_spline)
```

```python
y_spline_100 = spline_model(x[:100])
```

```python
plt.figure(figsize=(16, 5))
plt.plot(x[:100], y_spline_100, color='red', linewidth=2)
plt.scatter(x[:100], y[:100])
```
